# Feature Scaling

```python
X
```

```python
#Importing libraries for feature scaling
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
```

```python
#Initiating object for each library
standard_scalar = StandardScaler()
min_max_scalar = MinMaxScaler()
robust_scalar = RobustScaler()
```

```python
#Standard Scalar
X_scaled = pd.DataFrame(standard_scalar.fit_transform(X),columns=X.columns)
X_scaled.head()
```

```python
#MinMaxScalar
X_min_max = pd.DataFrame(min_max_scalar.fit_transform(X),columns=X.columns)
X_min_max.head()
```

```python
#Robust Scalar
X_robust = pd.DataFrame(robust_scalar.fit_transform(X),columns=X.columns)
X_robust.head()
```

```python
Y
```

# Model Building 

# 1}Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,plot_roc_curve
from sklearn.model_selection import cross_val_score,KFold
```

Without Feature Scaling

```python
#Hold Out method
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.30,random_state=0)
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_lr = lr.predict(x_test)
print(y_lr)
```

```python
#Evaluation Matrix
accuracy_lr = accuracy_score(y_test,y_lr)*100
print("Accuracy Score :",accuracy_lr)
print("Confusion Matrix :" ,confusion_matrix(y_test,y_lr))
print("Classification Report : ",classification_report(y_test,y_lr))
```

```python
#Cross val 
cross_lr = cross_val_score(lr,X,Y,cv=5)
cross_lr.mean()*100
```

```python
#Kfold
k_fold = KFold(5)
cross_k_fold = cross_val_score(lr,X,Y,cv=k_fold)
cross_k_fold.mean()*100
```

With Feature Scaling

```python
#Standard Scalar
x_train,x_test,y_train,y_test = train_test_split(X_scaled,Y,test_size=0.30,random_state=0)
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_lr = lr.predict(x_test)
print(y_lr)

#Evaluation Matrix
accuracy_lr = accuracy_score(y_test,y_lr)*100
print("Accuracy Score :",accuracy_lr)
print("Confusion Matrix :" ,confusion_matrix(y_test,y_lr))
print("Classification Report : ",classification_report(y_test,y_lr))

#Cross val 
cross_lr = cross_val_score(lr,X_scaled,Y,cv=5)
print(cross_lr.mean()*100)

#Kfold
k_fold = KFold(5)
cross_k_fold = cross_val_score(lr,X_scaled,Y,cv=k_fold)
print(cross_k_fold.mean()*100)
```

```python
#Min Max Scalar
x_train,x_test,y_train,y_test = train_test_split(X_min_max,Y,test_size=0.30,random_state=0)
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_lr = lr.predict(x_test)
print(y_lr)

#Evaluation Matrix
accuracy_lr = accuracy_score(y_test,y_lr)*100
print("Accuracy Score :",accuracy_lr)
print("Confusion Matrix :" ,confusion_matrix(y_test,y_lr))
print("Classification Report : ",classification_report(y_test,y_lr))

#Cross val 
cross_lr = cross_val_score(lr,X_min_max,Y,cv=5)
print(cross_lr.mean()*100)

#Kfold
k_fold = KFold(5)
cross_k_fold = cross_val_score(lr,X_min_max,Y,cv=k_fold)
print(cross_k_fold.mean()*100)
```

```python
#Robust Scalar
x_train,x_test,y_train,y_test = train_test_split(X_robust,Y,test_size=0.30,random_state=0)
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_lr = lr.predict(x_test)
print(y_lr)

#Evaluation Matrix
accuracy_lr = accuracy_score(y_test,y_lr)*100
print("Accuracy Score :",accuracy_lr)
print("Confusion Matrix :" ,confusion_matrix(y_test,y_lr))
print("Classification Report : ",classification_report(y_test,y_lr))

#Cross val 
cross_lr = cross_val_score(lr,X_robust,Y,cv=5)
print(cross_lr.mean()*100)

#Kfold
k_fold = KFold(5)
cross_k_fold = cross_val_score(lr,X_robust,Y,cv=k_fold)
print(cross_k_fold.mean()*100)
```
